```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

import plotly.express as px

import dash
import dash_html_components as html
import dash_bootstrap_components as dbc
import dash_core_components as dcc
from dash.dependencies import State ,Output,Input
```

```python
iris_df = pd.DataFrame(load_iris()['data'])
iris_df.columns= load_iris()['feature_names']
iris_df

```

```python
def km_cluster(n):
    km = KMeans(n_clusters=n)
    km.fit(iris_df.values)    
    return km.labels_,km.cluster_centers_
```

```python

app = dash.Dash(__name__)
app.layout = html.Div(
    [
        
        dcc.Input(
            id = 'n-input',
            type = 'number'
        ),
        html.Button(
            'Submit!',
            id = 'submit-btn',
            n_clicks= 0
        ),
        dcc.Dropdown(
            id = 'drop-x',
            options = [{'label':i,'value':i } for i in iris_df.columns],
            value = [iris_df.columns[0]]
        ),
        dcc.Dropdown(
            id = 'drop-y',
            options = [{'label':i,'value':i } for i in iris_df.columns],
            value = [iris_df.columns[1]]
        ),

        dcc.Graph(
            id = 'scatter-plot'
        )
        
        
    ]
)
@app.callback(
    Output('scatter-plot','figure'),
    Input('submit-btn','n_clicks'),
    State('n-input','value'),
    State('drop-x','value'),
    State('drop-y','value'),
    
)
def update_scatter(nBtn,n,first,second):    
    tmp = iris_df.copy()
    tmp['cluster'], center = km_cluster(n)
    tmp['cluster'] ='cluster - ' + tmp['cluster'].astype(str)  
    center = pd.DataFrame(center)
    center.columns = iris_df.columns
    center['cluster'] = 'center'
    tmp = tmp.append(center, ignore_index=True)
    fig = px.scatter(tmp, x = first, y = second, color="cluster")
    # fig = px.density_contour(tmp, x = first, y = second, color="cluster")
    return fig


@app.callback(
    Output('drop-x',"options"),
    Input('drop-y',"value")
)
def check_y(y):
    return [{'label':i,'value':i, 'disabled': i == y } for i in iris_df.columns]

@app.callback(
    Output('drop-y',"options"),
    Input('drop-x',"value")
)
def check_x(x):
    return [{'label':i,'value':i ,'disabled' : i == x } for i in iris_df.columns]

app.run_server()







```
