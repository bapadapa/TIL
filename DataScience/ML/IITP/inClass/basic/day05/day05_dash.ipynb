```python
import numpy as np
import pandas as pd
# pip install dash
# pip install dash_bootstrap_components

import dash
import dash_html_components as html
import dash_core_components as dcc
from dash.dependencies import Input, Output

import plotly.express as px
```

```python
# app 생성
df = px.data.gapminder()
app = dash.Dash(__name__)
app.layout = html.Div(
    [
        html.H1('my first dash_app'),
        html.H2('my second dash_app'),
        dcc.Dropdown(
            id = 'my-first-dropdown',
            options = [{'label':str(i)+'년','value':i} for i in df.year.unique()]   
        ),
        dcc.Graph(
            id = "my-first-grape"            
            # id = "my-first-grape",            
            # figure = fig
        ),
        dcc.Dropdown(
            id = 'my-second-dropdown',
            options = [{'label' : i ,'value': i } for i in df.continent.unique()]   
        ),
        html.Hr(),
        dcc.Dropdown(
            id  = 'my-third-dropdown'
        )
    
    ]
)

@app.callback(    
    # Output( ID, Value )
    Output('my-first-grape','figure'),
    # Input( ID , Value )
    Input('my-first-dropdown', 'value')
)
def update_grape(year):
    qry = "year == "+ str(year)
    return  px.scatter(df.query(qry) , x ="gdpPercap",y= "lifeExp", size ="pop",color ="continent",hover_name="country",log_x=True,size_max = 60)
    
    

@app.callback(
    Output('my-third-dropdown', 'options'),
    Input('my-second-dropdown','value')
)
def update_my_third_dropdown(continent):
    return [{'label' :i ,'value': i } for i in df.loc[df.continent == continent,'country'].unique()]     
```

```python
app.run_server(port='1234')
```
