### odds

```python
from matplotlib import pyplot as plt

def xrange(start, end, interval):
    numbers = []
    while start < end:
        numbers.append(start)
        start += interval
    return numbers

x = xrange(0.001, 1.0, 0.001)
y = []

for p in x:
    y.append(p / (1-p))

plt.plot(x,y)
plt.ylim([0,20])
plt.xlabel('probability')
plt.ylabel('odds')
plt.show()
```

### logit

```python
import numpy as np
logit = []
for p in x:
    logit.append(np.log(p / (1-p)))

plt.plot(x,logit)
plt.xlabel('probability')
plt.ylabel('logit')
plt.show()
```
