```python
import numpy as np
import pandas as pd

import plotly.express as px

import dash
import dash_html_components as html
import dash_core_components as dcc
import dash_bootstrap_components as dbc
from dash.dependencies import Input,Output,State
```

```python
df = px.data.gapminder()
```

```python
df.loc[df['continent']=='Asia' , 'country'].unique()
```

```python
conti_radio = dcc.RadioItems(
            id = 'continent-radio',
            options =[
                # {'label':,'value':}
                {'label': i,'value': i } for i in df.continent.unique()
            ]
        )
c_dropdown =dcc.Dropdown(
            id = 'country-dropdown',
            options = [{'label':i,'value':i} for i in df.country.unique()]
        )
y_slider =dcc.RangeSlider(
            id = 'year-slider',
            min= df.year.min(),
            max= df.year.max(),
            marks = {int(i) : str(i)+'년도' for i in df.year.unique()}
        )
p_graph = dcc.Graph(
            id = 'bar-graph'
        )

control = dbc.Card(
    [
        dbc.CardBody(
            [
                
                dbc.Card(
                    [dbc.CardHeader('대륙을 선택하세요'),
                    conti_radio]
                    ),
                    ],style = {'background-color':'black'}
        ),
        dbc.CardBody(
            [
                dbc.Card(
                    [
                    dbc.CardHeader('나라을 선택하세요'),
                    c_dropdown]
                    ),
            ],style = {'background-color':'black'}
        ),
        dbc.CardBody(
            [
                dbc.Card(
                    [dbc.CardHeader('연도을 선택하세요'),
                    y_slider]
                    )
            ],style = {'background-color':'black'}
        )
    ]
)
graph= dbc.Card(p_graph)
main = dbc.CardDeck(
    [
        control,
        graph
        
    ]
)
```

```python
#  데이터를 여러개 받으려면, list형태로 받아야한다.
app = dash.Dash(__name__,external_stylesheets=[dbc.themes.SOLAR])
app.layout = html.Div(
    [   
        main
        
    ]
)
```

```python
@app.callback(
    Output('country-dropdown','options'),
    Input('continent-radio', 'value')
)
def updateCountryDropdown(continent):
    return [{'label':i,'value':i} for i in df.loc[df['continent']==continent , 'country'].unique()]


@app.callback(
    Output('range-value','children'),
    Input('year-slider','value')
)
def updateRangeValue(year):
    return html.H6('선택한 년도는{}년 ~ {}년 입니다.'.format(str(year[0]),str(year[1])))

@app.callback(
    Output('bar-graph','figure'),
    Input('country-dropdown','value'),
    Input('year-slider','value')
)
def updateBarGraph(country,year):
    return px.bar(df.loc[(df.country ==country) & ((df.year >=int(year[0])) & (df.year <= int(year[1]))) ,['year','pop']] ,x = 'year',y = 'pop',color = 'pop')
```

```python
app.run_server()
```
