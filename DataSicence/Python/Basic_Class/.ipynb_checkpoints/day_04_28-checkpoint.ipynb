# 반복문

## While문

### 기본 개념
    - for문과 같이 반복문이다.
    - for문은 제한된 범위내에서 반복을 하지만, while문은 탈출문을 작성하여 반복을 멈춘다
    - 위와같은 이유로, for문은 반복횟수를 알고있을 때 많이 사용하고, while문은 특정조건에 멈춰야 할 때 주로 사용한다
   
### 사용 방법
    while 정지조건 : 
        연산 ( 명령줄)
    

```python
# while 간단 예시 1
i = 0
sum = 0
while i < 100 :
    sum += i
    i+=1
print (sum)
```

```python
# while 간단 예시 2

import random
random_list = []
while len(random_list) < 1001:
    x = random.randint(0,1000)
    if x in random_list:
        pass
    else:
        random_list.append(x)
print(random_list)
```

```python
##나는 행복합니다를 출력하는 함수
def first_func(name):
    print(name,"는 행복합니다.")    
##값을 입력받아 그 값의 제곱을 출력하는 함수
def second_func(x):
    return x**2
```

```python
first_func('승현')
second_func(2)
```

```python
dic = {}
for k,v in enumerate(range(3,10001,3)):
    if k %3 == 0:
        dic[v] = second_func(v)
dic
```

```python
 #로또 만들기
def lotto (n1, n2, cnt):
    out = []
    while len(out) < cnt:        
        num = random.randint(n1,n2)
        if num in out :
            pass
        else:
            out.append(num)
    return out
```

### map함수
-map([함수 / 자료형] , [ 데이터 객체 ] ) 


```python
# 방법 1
x = ['34.5','35.6']
x1 = float(x[0])
x2 = float(x[1])
l1 = [x1,x2]
print(l1)

# 방법 2
l2 = list(map(float,x))
print(l2)

## 함수 삽입하여 사용하기
def myReplace(x):
    return x.replace('.', '')
l3 = list(map(myReplace,x))
print(l3)
```

```python
#문장을 읽어드리기
sentence = "지금까지 변수를 만들고 input 함수를 사용해보았습니다. 변수는 값이나 계산 결과를 저장할 때 사용한다는 점, 변수를 만드는 방법, 변수 이름을 짓는 방법만 기억하면 됩니다. 특히 input과 split의 결과가 문자열라는 점이 중요합니다. 따라서 숫자 계산을 한다면 int, float를 사용해서 결과를 숫자로 변환해주어야 한다는 점 기억하세요. 그리고 split의 결과를 모두 int, float로 변환할 때는 map을 사용하면 편리합니다."


#문장을 리스트 만들기
## 나열 형식으로 구현하기
def wordCnt(sentence):
    sentenceList = list(sentence.split())
    sentenceListLen = []
    for i in sentenceList:
        sentenceListLen.append(len(i))
    return sentenceListLen


## 맵을 이용하여 구현하기
def wordCnt2(sentence):
    return list(map(len, list(sentence.split())))


ans = wordCnt2(sentence)
print(ans)
```

# 조건문

## if..elif..else

- if문
    - if 조건 : 
        - 연산
    - elif 조건 :
        - 연산 :
    - else :
        - 연산

- if문은 if -> elif -> ... -> elif -> else 순으로 실행된다.
    - 만약 현재 위치보다 위의 조건에서 True 즉, 연산을 수행했다면, 조건문 실행을 종료한다.


```python
## 카와 몸무계를 입력받아 bmi를 구하고 bmi가 
# 25보다 크면 A
# 20~25 면 B
# 20보다 작으면 C를 출력하시오
height , weight = map(int,input().split())
height /= 100
bmi = weight /(height**2)
x = None
if bmi > 25 :
    x = 'A'
elif bmi > 20 :
    x = 'B'
else:
    x = 'C'
print(x)
```

```python
# List를 입력받아 길이가 0보다 크면 길이를 출력하고 그렇지 않는다면 길이가 작습니다를 출력
# 값중에서 가장 큰 값이 10보다 작으면 'Small' , 그렇지 않으면 'Large'를 출력하시오
i = list(range(-10 , 10))
if len(i) > 10:
    print(len(i))
    if max(i) < 10:
        print("Small")
    else : 
        print("Large")
else : 
    print("길이가 작습니다")
```
